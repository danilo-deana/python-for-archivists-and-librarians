# Il formato JSON

I documenti in formato json sono utilizzati per scanbiare dati tra diverse applicazione. La struttura di un documento JSON.

Esiste una specifica libreria chiamata json che permette di leggere e manpolare documenti in formato json.

# Lettura

```python
import json

with open('json/ArchiviAllxOD.json') as documento:
  dati = json.load(documento)

print(dati)
```

# Struttura

```python
import json

with open('json/ArchiviAllxOD.json') as documento:
  dati = json.load(documento)

for chiave0 in dati:
    print(chiave0)
```

```python
valore0 = dati[chiave0]

type(valore0)
```

```python
for chiave1 in valore0:
    print(chiave1)
```

```python
valore01 = valore0['request']

type(valore01)
```

```python
valore03 = valore0['meta-info']

type(valore03)
```

```python
valore02 = valore0['SelArchivi']

type(valore02)
```

```python
for elemento in valore02:
    for chiave3 in elemento:
        print(chiave3)
    break
```

# Modifiche

```python
for indice, valore in enumerate(valore02):
    dati['ArchiviAllxOD']['SelArchivi'][indice]['RETE_DOCUMENTARIA'] = 
        'Sistema bibliotecario provinciale di Milano'
```

# Scrittura

```python
import json

with open('json/ArchiviAllxOD_PyAL.json', 'w') as destinazione: 
    json.dump(dati, destinazione, indent = 4)
```
