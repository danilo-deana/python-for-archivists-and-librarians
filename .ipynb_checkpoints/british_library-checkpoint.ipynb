# Elenco dei documenti contenuti nella cartella

```python
import os

documenti = []

for root, directories, files in os.walk('british_library/'):
    for file in files:
        documenti.append(os.path.join(root, file))
        

documenti.sort()

print(documenti)
```

# Registrazioni bibliografiche

```python
import pandas as pd

dfRCD = pd.read_csv('british_library/records.csv')

dfRCD.info()
```

```python
pd.set_option('display.max_columns', None)

dfRCD.head()
```

# Distribuzione dei documenti per data di pubblicazione

```python
def anno(date):
    if type(date) == float:
        return 0
    if date[0:4].isnumeric():
        return int(date[0:4])
    return 0
    
dfRCD['year'] = dfRCD.apply(lambda x: anno(x['Date of creation/publication']), axis = 1)

dfTMP = dfRCD[dfRCD['year'] != 0]

dfTMP.hist(column = 'year', figsize = (11.7, 8.3))
```

```python
dfTMP['year'].describe()
```

```python
dfRCD.sort_values(by = ['year'], inplace = True)

dfTMP = dfRCD[dfRCD['year'] != 0]

dfTMP.head()
```
